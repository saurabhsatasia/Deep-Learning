## Kaggle Pima Diabetes Prediction using ANN with Pytorch ON CPU & GPU

```python
import torch
import pandas as pd
import numpy as np
import seaborn as sns

print(torch.__version__)
print(torch.cuda.is_available())
print(torch.cuda.get_device_name())
print(torch.cuda.current_device())
print(torch.cuda.memory_cached())
print(torch.cuda.memory_allocated())
```

```python
df = pd.read_csv("diabetes.csv")

df.sample(5)
```

```python
df['Outcome_sns'] = np.where(df['Outcome']==1, "Diabetic", "Non Diabetic")
sns.pairplot(df.drop("Outcome", axis=1), hue="Outcome_sns")
```

```python
from sklearn.model_selection import train_test_split

X=df.drop(["Outcome", "Outcome_sns"], axis=1).values
y=df['Outcome'].values

X_train,X_test, y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=999)
```

### Using CPU

```python
# import pytorch libs
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
# 1. Create Tensors --> Inputs have to be created as Tensors
## Independent features always have to be converted to Float Tensor

X_train = torch.FloatTensor(X_train)
X_test = torch.FloatTensor(X_test)

y_train = torch.LongTensor(y_train) # long tensor will not convert to floating dtype
y_test = torch.LongTensor(y_test)
```

```python
df.shape
```

```python
# 2. Creating ANN model
# Categorical cross entropy loss function to get probability values
# init requires input features, hidden layers, output layer
class ANN_Model(nn.Module):
    def __init__(self, input_feature=8, hidden1=20, hidden2=20, out_features=2):
        super().__init__()
        self.ful_connected1 = nn.Linear(input_feature, hidden1)
        self.ful_connected2 = nn.Linear(hidden1, hidden2)
        self.out_layer = nn.Linear(hidden2, out_features)
    
    def forward(self, x): # x will track the GD during BPP
        x = F.relu(self.ful_connected1(x))
        x = F.relu(self.ful_connected2(x))
        x = self.out_layer(x)
        return x
```

```python
# instantiate ANN_Model
torch.manual_seed(20) # set a seed to apply the initial weights

model = ANN_Model()
model.parameters # like model.summary in Keras
```

```python
# For BPP, define a Loss Funtion and Optimizer
loss_function = nn.CrossEntropyLoss() # can also use for multiclass-classificaiton
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
epochs = 500
final_losses = []

for i in range(epochs):
    i=i+1
    y_pred = model.forward(X_train)
    
    loss = loss_function(y_pred, y_train)
    final_losses.append(loss)
    
    if i%10 == 1:
        print(f"Epoch number: {i} and the Loss: {loss.item()}")
        
    # Reduxe the loss
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
# plot the loss function
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(range(epochs), final_losses)
plt.ylabel("Loss")
plt.xlabel("Epoch")
```

```python
# Prediction

predictions=[]

with torch.no_grad():
    for i,data in enumerate(X_test):
        y_pred = model(data)
        predictions.append(y_pred.argmax().item())
        print(y_pred.argmax().item())
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predictions)
print(cm)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(cm, annot=True)
plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, predictions)*100
```

## Using GPU

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
##### Creating Tensors
X_train=torch.FloatTensor(X_train).cuda()
X_test=torch.FloatTensor(X_test).cuda()
y_train=torch.LongTensor(y_train).cuda()
y_test=torch.LongTensor(y_test).cuda()
```

```python
X_train.device
```

```python
#### Creating Modelwith Pytorch

class ANN_Model(nn.Module):
    def __init__(self,input_features=8,hidden1=20,hidden2=20,out_features=2):
        super().__init__()
        self.f_connected1=nn.Linear(input_features,hidden1)
        self.f_connected2=nn.Linear(hidden1,hidden2)
        self.out=nn.Linear(hidden2,out_features)
    def forward(self,x):
        x=F.relu(self.f_connected1(x))
        x=F.relu(self.f_connected2(x))
        x=self.out(x)
        return x
```

```python
torch.manual_seed(20)
model=ANN_Model()
```

```python
model.parameters
```

```python
model=model.cuda()
```

```python
for i in model.parameters():
    print(i.is_cuda)
```

```python
###Backward Propogation-- Define the loss_function,define the optimizer

loss_function=nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(model.parameters(),lr=0.01)
```

```python
import time
start_time=time.time()
epochs=10000
final_losses=[]
for i in range(epochs):
    i=i+1
    y_pred=model.forward(X_train)
    loss=loss_function(y_pred,y_train)
    final_losses.append(loss)
    if i%10==1:
        print("Epoch number: {} and the loss : {}".format(i,loss.item()))
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
print(time.time()-start_time)
```

```python
### plot the loss function
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(range(epochs),final_losses)
plt.ylabel('Loss')
plt.xlabel('Epoch')
```

```python
#### Prediction In X_test data
predictions=[]
with torch.no_grad():
    for i,data in enumerate(X_test):
        y_pred=model(data)
        predictions.append(y_pred.argmax().item())
        print(y_pred.argmax().item())
```
