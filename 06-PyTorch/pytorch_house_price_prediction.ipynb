```python
# Categorical Features
cat_features = ["MSSubClass", "MSZoning", "Street", "LotShape"]
out_feature = "SalePrice"

from sklearn.preprocessing import LabelEncoder

lbl_encoders = {}
for feature in cat_features:
    lbl_encoders[feature] = LabelEncoder()
    df[feature] = lbl_encoders[feature].fit_transform(df[feature])
    
df.sample(5)
```

### Stacking and Converting Categorical Features into Tensors

```python
cat_features=np.stack([df['MSSubClass'], df['MSZoning'], df['Street'], df['LotShape']], axis=1)
print(cat_features)
print("-"*60)


cat_feat_tensor = torch.tensor(cat_features, dtype=torch.int64)
print(cat_feat_tensor)    
```

### Stacking and Converting Continous Features into Tensors

```python
# create continous variable
cont_features=[]
for i in df.columns:
    if i in ["MSSubClass","MSZoning","Street","LotShape", "SalePrice"]:
        pass
    else:
        cont_features.append(i)
        
print(cont_features)
```

```python
# stacking continous variables to a tensor
cont_values = np.stack([df[i].values for i in cont_features], axis=1)
print(cont_values)
print("-"*60)

cont_feat_tensor = torch.tensor(cont_values, dtype=torch.float)
print(cont_feat_tensor)
print("-"*60)
print(cont_feat_tensor.dtype)
```

```python
# Target/Dependent Feature
y = torch.tensor(df['SalePrice'].values, dtype=torch.float).reshape(-1,1)
print(y)
```

```python
print(f"Categorical Features Shape: {cat_feat_tensor.shape}")
print(f"Continous Features Shape: {cont_feat_tensor.shape}")
print(f"Target/Dependent Feature Shape: {y.shape}")
```

### Embedding Layer w.r.t. Categorical Features
Embedding requires Label Encoded values for Categorical Features.
`Here the length/dimension of a categorical feature is important, it basically says that How many INPUTS , the embedding layer should have and based on that How many OUTPUTS should be created.`

```python

print(len(df['MSSubClass'].unique()))

# embedding the size/length for Categorical comlumns
cat_dims = [len(df[col].unique()) for col in ["MSSubClass","MSZoning","Street","LotShape"]]
print(f"Length of each Categorical Features: {cat_dims}")

# Thumb rule for the output dimension: 
## Output dimension should be set based on Input dimension, 
### it should be `min(50, feature_dimension/2)`
embedding_dim = [(x, min(50, (x+1)//2)) for x in cat_dims]
print(f"Embedded Dimesions : {embedding_dim}")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

embeded_representation = nn.ModuleList([nn.Embedding(inp , out) for inp,out in embedding_dim])
print(embeded_representation)
```

```python
print(cat_feat_tensor)
```

```python
embedding_val=[]
for i,e in enumerate(embeded_representation):
    embedding_val.append(e(cat_feat_tensor[:,i]))
    
print(embedding_val)
```

```python
# above tensor is not stacked column-wise. Stacking should happen column-wise
# use concatenation to convert labels into vector

z = torch.cat(embedding_val, 1) # axis=1
print(z)
```

```python
# implement dropout to handle overfitting problem
drop_out = nn.Dropout(0.4) # 40% values are dropped off
final_embeded = drop_out(z)

print(final_embeded)
```

```python
# Create Feed Forward Neural Network
import torch
import torch.nn as nn
import torch.nn.functional as F

class FeedForwardNN(nn.Module):
    def __init__(self, embedding_dim, n_cont, out_sz, layers, p=0.5): #p is dropout ratio
        super().__init__()
        self.embeds=nn.ModuleList([nn.Embedding(inp,out) for inp,out in embedding_dim])
        self.emb_drop = nn.Dropout(p)
        self.bn_cont = nn.BatchNorm1d(n_cont) # batch normalization w.r.t. continous features
        
        layerlist = []
        n_emb = sum((out for inp,out in embedding_dim))
        n_in = n_emb + n_cont
        
        for i in layers:
            layerlist.append(nn.Linear(n_in, i))
            layerlist.append(nn.ReLU(inplace=True))
            layerlist.append(nn.BatchNorm1d(i))
            layerlist.append(nn.Dropout(p))
            n_in = i
        layerlist.append(nn.Linear(layers[-1], out_sz))
        self.layers = nn.Sequential(*layerlist)
        
    def forward(self, x_cat, x_cont):
        embeddings=[]
        for i,e in enumerate(self.embeds):
            embeddings.append(e(x_cat[:,i]))
        x = torch.cat(embeddings, 1)
        x = self.emb_drop(x)
        
        x_cont = self.bn_cont(x_cont)
        x = torch.cat([x,x_cont], 1)
        x = self.layers(x)
        return x
```

```python
len(cont_features)
```

```python
torch.manual_seed(100)
model = FeedForwardNN(embedding_dim=embedding_dim, n_cont=len(cont_features), out_sz=1, layers=[100,50], p=0.4)
model
```

```python
### Define Loss and Optimizer

loss_function = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
print(df.shape)
print(cont_feat_tensor)
print(cont_feat_tensor.shape)

batch_size = 1200
test_size = int(batch_size*0.15) # 15% of the total batch_size --> 1200*0.15=180records as test data

train_categorical = cat_feat_tensor[: batch_size-test_size] # :,1200-180
test_categorical = cat_feat_tensor[batch_size - test_size : batch_size] # 1200-180 : 1200

train_cont = cont_feat_tensor[:batch_size - test_size] # :1200-180
test_cont = cont_feat_tensor[batch_size-test_size : batch_size] #1200-180 : 1200

y_train = y[:batch_size-test_size] # : 1200-180
y_test = y[batch_size - test_size : batch_size] # 1200-180 : 1200
```

```python
print(f"Length of Train Categorical: {len(train_categorical)}")
print(f"Length of Test Categorical: {len(test_categorical)}")
print("-"*60)

print(f"Length of Train Continous: {len(train_cont)}")
print(f"Length of Test Continous: {len(test_cont)}")
print("-"*60)

print(f"Length of Train Target: {len(y_train)}")
print(f"Length of Test Target: {len(y_test)}")
```

```python
epochs=5000
final_losses=[]
for i in range(epochs):
    i=i+1
    y_pred=model(train_categorical,train_cont)
    loss=torch.sqrt(loss_function(y_pred,y_train)) ### RMSE
    final_losses.append(loss)
    if i%10==1:
        print("Epoch number: {} and the loss : {}".format(i,loss.item()))
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(range(epochs), final_losses)
plt.ylabel('RMSE Loss')
plt.xlabel('epoch');
```
